* ## Confounding Variables


>1.   Confounding variables are external factors that influences both independent and dependent variables which may cause a misleading association between them.  

>2.   Potential confounding variables in MxMH dataset:  

>> *   Hours per day & Insomnia - People with insomina may listen to more music at night. This may cause the confusion if listening more music is causing insomnia or if insomnia results in more listening hours. This confounds the result between music effects and mental health.  

>> *   Hours per day, Instrumentalist & Composer - People who play instruments or compose music may listen their music multiple times to practice or refine it thus increasing listen hours. This confounds the result between music effects and mental health.

>3. Dealing with counfounding variables:  

>> *   Stratification - Spilt the datatsets into subgroups based on confounders - Example: Different mental health challenges, Instrumentalist vs Non-Instrumentalist, Composer vs Non-composer.

```python
# Calculating z-scores to identify outliers for numeric columns
def outliers(df):
  numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
  print(numeric_cols)
  z_scores = (df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std()
  abs_z_scores = abs(z_scores)
  outliers = (abs_z_scores > 3).any(axis=1)
  outlier_rows = df[numeric_cols][outliers]
  outliers_count = outliers.sum()
  print(f'Number of outliers: {outliers_count}')
  return outlier_rows

outlier_rows = outliers(df)
print(outlier_rows)
```

# Data Cleaning Plan
> **Replacing Null Values:**


*   age - Since only 1 value is missing and the column is numeric, we can replace the value with median.
*   primary_streaming_device - Since only 1 value is missing and the coulumn is categorical, we can replace the value with the mode.
*   listening_while_working - Since 3 values are missing and the coulumn is categorical, we can replace the values with the mode.
*   instrumentalist - Since 4 values are missing and the coulumn is categorical, we can replace the values with the mode.
* composer - Since only 1 value is missing and the coulumn is categorical, we can replace the value with the mode.
* foreign_languages - Since 4 values are missing and the coulumn is categorical, we can replace the values with the mode.
* bpm_fav_genre - it has 107 missing values, the missing values can be replaced with the median bpm of the corresponding genre. bpm has a range of 30-400 in the real world, thus any value which is missing or is out of range will be replaced by the bpm median of the corresponding genre. Out of range values will be replaced by Nan and then replaced with median.
* music_effects - it has 8 missing values. Since its the target variable and the column is categorical, we will remove it from the original dataset, and save the 8 rows as a different dataframe under a new category 'Not responded' which will be anaylzed later.

> **Fixing Data Type Inconsistency**


*  All the columns which have missing values have data type inconsistency as NaN is considered as a float. Once all the null values are replaced, there should be no data type inconsisency.

> **Handling Outliers**
* bpm_fav_genre has one value of 999999999.0 which is an unrealistic outlier. This will be replaced by the bpm median of the corresponding genre.
* While handling missing values in bpm, all the values out of range (30-400) will be replaced by Nan and replaced with bpm median.
* hours_per_day have values as 24 which is unrealistic outlier. All values above 20 will be capped at 20.






# Data Cleaning

```python
# Fixing the null values
import numpy as np
df['age']= df['age'].fillna(df['age'].median())
df['primary_streaming_service']= df['primary_streaming_service'].fillna(df['primary_streaming_service'].mode()[0])
df['listening_while_working']=df['listening_while_working'].fillna(df['listening_while_working'].mode()[0])
df['instrumentalist']= df['instrumentalist'].fillna(df['instrumentalist'].mode()[0])
df['composer']= df['composer'].fillna(df['composer'].mode()[0])
df['foreign_languages']= df['foreign_languages'].fillna(df['foreign_languages'].mode()[0])
df.loc[(df['bpm_fav_genre'] < 30) | (df['bpm_fav_genre'] > 400), 'bpm_fav_genre'] = np.nan
df['bpm_fav_genre'] = df.groupby('fav_genre')['bpm_fav_genre'].transform(lambda x: x.fillna(x.median()))
df_music_effects_null = df[df['music_effects'].isnull()]
df_music_effects_null.loc[:, 'music_effects'] = df_music_effects_null['music_effects'].fillna('Not responded')
df_cleaned = df[~df['music_effects'].isnull()].copy()
print('Checking for null values:\n', df_cleaned.isnull().sum())
print('\nNot responded music effects dataset', df_music_effects_null.head(5))
```

# Feature Engineering

*   Firstly, we will drop columns which are not significant to the target variable - timestamp and permissions.
*   Next, We are creating new features -
> mental_health_score - Since all the individual health scores are positively correlated, all the scores are combined to represent person's overall mental health indicator.  
> music_engagement - People who play instruments, compose music, listen and explore foreign langugae music and higher engagement which is combined as a single indicator.

*   Next, we will be applying label encoding to categorical columns for easier modelling.
*   Finally, we will use mutual information score to rank features based on non linear depedency on target variable and drop the features which are unimportant.

# Extra credit - Feature Engineering
*   Backward elimitaion - we will start with all 24 features and iteratively drop the leat important feature.



```python
# Feature Engineering - backward elimintaion
def backward_elimination(X, y , min_features):
  features = list(X.columns)
  history = []
  while len(features) > min_features:
    X_train, X_test, y_train, y_test = train_test_split(X[features], y, test_size=0.2, random_state=42, stratify=y)
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train)
    y_pred = rf_model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    importances = pd.Series(rf_model.feature_importances_, index=features).sort_values(ascending=False)
    least_important_feature = importances.idxmin()
    history.append({
        'num_features': len(features),
        'least_important_feature': least_important_feature,
        'accuracy': accuracy
    })
    features.remove(least_important_feature)
  history_df = pd.DataFrame(history)
  return features, history_df
```

```python
final_features, history_df = backward_elimination(X_mi, y_mi, min_features=15)
print('Selected features after backward elimination:', final_features)
print('\n Elimiated History \n', history_df)
```

# Algorithm  Selection
1.   Assumptions of Random Forest -
> *  No assumption about data distribution - Random forest does not require features to be normally distributed.
> *   No assumption for linearity - Random forest captures non linear relationship between features and target.
> * Random Forest are less sensitive to outliers, as averaging many decision trees reduces the impact of outliers.

2.  Issues & Mitigations
> * Individual decision trees can overfit easiy but averaging many trees in Random Forest avoids overfitting.
> * Overfitting can be further reduced by tuning max_depth, min_samples_spilt, min_samples_leaf, n_estimators.

3. Random Forest Good Candidacy
> * Music behavious and mental health effects are non-linear and complex and Random Forest captures these patterns better than linear models.
> * Radom Forest handles both numerical and categorical columns (after label encoding)
> * Random Forest uses random subsets of features at each spilt and helps in identifying feature importance which helps in backward elimiation.
> * The target variable - music_effects has multiple categories which is handled directly by Random Forest.







# Hyperparameter Selection
Random forest has 5 hyperparameters -
1.   n_estimators - Number of trees in the forest
2.   max_depth - Maximum depth of each decision tree
3.   min_samples_spilt - Minimum number of samples required to spilt a node
4.   min_samples_leaf - Minimum number of samples required at a leaf node
5.   max_features - Number of features considered for spiltting at each node


For tuning, **Grid Search** can be used which test combinations of hyperparameters and gives the best parameters based on some decided scoring (like accuracy).


# Post-processing
Unlike a single decision tree, Random Forest does not require post pruning since after each tree is built, the ensemble averages the predictions which avoids overfitting.

```python
reports = {
    'Random Forest': classification_report(y_test_encoded, y_pred_tuned, output_dict=True, zero_division=0),
    'Logistic Regression': classification_report(y_test_encoded, y_pred_log, output_dict=True, zero_division=0),
    'SVM': classification_report(y_test_encoded, y_pred_svm, output_dict=True, zero_division=0),
    'XGB': classification_report(y_test_encoded, y_pred_xgb, output_dict=True, zero_division=0)
}
class_labels = {i: name for i, name in enumerate(le.classes_)}
rows = []
for model_name, report in reports.items():
  row = {
      'Model': model_name,
      'Accuracy': round(report['accuracy'], 3),
      'F1 (Weighted)': round(report['weighted avg']['f1-score'], 3),
      'F1 (Macro)': round(report['macro avg']['f1-score'], 3)
  }
  for num_label, class_name in class_labels.items():
        recall = report.get(num_label, report.get(str(num_label), {})).get('recall', 0)
        row[f'Recall ({class_name})'] = round(recall, 2)
  rows.append(row)
metric_df = pd.DataFrame(rows)
metric_df = metric_df.set_index('Model').sort_values(by='Accuracy')
print(metric_df)
```

From the metric above we can conclude that XGB is the optimal model amongst the four models as XGB achieves the best balance between accurcay, F1 and recall.
MxMH dataset is small dataset with high class imbalance, 'Worsen' has only 3 rows. Thus high accuracy means high variance and overfitting of data and accurcay cannpt be used alone for models evalutaion. Reacll measures how many actual cases of each data are correctly idenitifed, while F1 score balances precision and recall to evaluate reliable performance across all classes, making them more informative than accurcay for our dataset.
